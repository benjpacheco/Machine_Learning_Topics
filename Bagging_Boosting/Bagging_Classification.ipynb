```python
X = df_heart.loc[:, df_heart.columns != 'HeartDisease']
y = df_heart['HeartDisease']
```

```python
dummies = pd.get_dummies(X[['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope']])
X = pd.concat([X, dummies], axis=1)
X.drop(['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope'], axis=1,inplace=True)
X.head()
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

model_params = {
    'svm': {
        'model': SVC(gamma='auto'),
        'params': {
            'svm__C': [1, 10, 20],
            'svm__kernel': ['rbf', 'linear']
        }
    },
    'tree': {
        'model': DecisionTreeClassifier(),
        'params': {
            'tree__criterion': ['gini', 'entropy'],
            'tree__splitter': ['best', 'random'],
            'tree__max_depth': range(1, 20)
        }
    }
}
```

```python
from sklearn.pipeline import Pipeline
scores = []

for model_name, mp in model_params.items():
    pipe = Pipeline([
        ('scaler', StandardScaler()),
        (model_name, mp['model'])
    ])
    clf = GridSearchCV(estimator = pipe, param_grid = mp['params'], cv=5, return_train_score=False)
    clf.fit(X, y)
    scores.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })
```

```python
df = pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])
df
```

```python
# bagged models
model_params = {
    'svm': {
        'model': BaggingClassifier,
        'kwargs': {'base_estimator': SVC(), 'n_estimators': 100, 'max_samples': 0.8, 'oob_score': True,
                   'random_state': 0},
        'params': {
            'svm__base_estimator__gamma': ['auto'],
            'svm__base_estimator__C': [1, 10, 20],
            'svm__base_estimator__kernel': ['rbf', 'linear']
        }
    },
    'tree': {
        'model': BaggingClassifier,
        'kwargs': {'base_estimator': DecisionTreeClassifier(), 'n_estimators': 100, 'max_samples': 0.8, 'oob_score': True,
                   'random_state': 0},
        'params': {
            'tree__base_estimator__criterion': ['gini', 'entropy'],
            'tree__base_estimator__splitter': ['best', 'random'],
            'tree__base_estimator__max_depth': range(1, 20)
        }
    }
}
```

```python
bag_scores = []

for model_name, mp in model_params.items():
    pipe = Pipeline([
        ('scaler', StandardScaler()),
        (model_name, mp['model'](**mp['kwargs']))
    ])
    clf = GridSearchCV(estimator = pipe, param_grid = mp['params'], cv=5, return_train_score=False)
    clf.fit(X, y)
    bag_scores.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })
```

```python
df_bag = pd.DataFrame(bag_scores, columns=['model', 'best_score', 'best_params'])
df_bag
```

Before bagging the Support Vector Machine scored higher in performance compared to Decision Trees. However after bagging we see that the ensemble method Decision Tree scored 83.3% compared to SVM 82.9%. This concludes that bagging is a better fit for ensemble methods.
