### Excercise

Predict wether the person has heart disease or not.

1. Load heart disease into pandas df
2. Remove outliers using Z score. Usual guildeline is to remove anything that has Z score > 3 or Z score < -3
3. Convert text columns to numbers using label encoding and one hot encoding
4. Apply scaling
5. Build a classification model using various methods (SVM, logistic regression, random forest) and check which model gives you the best accuracy
6. Now use PCA to reduce dimensions, retrain your model and see what impact it has on your model in terms of accuracy. Keep in mind that many times doing PCA reduces the accuracy but computation is much lighter and that's the trade off you need to consider while building models in real life.

```python
df_heart = pd.read_csv('heart.csv')
df_heart.head()
```

```python
df_heart.describe()
```

```python
from scipy import stats

df_heart = df_heart[(np.abs(stats.zscore(df_heart.select_dtypes(include=np.number))) < 3).all(axis=1)]
df_heart.describe()
```

```python
X = df_heart.loc[:, df_heart.columns != 'HeartDisease']
y = df_heart['HeartDisease']
```

```python
dummies = pd.get_dummies(X[['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope']])
X = pd.concat([X, dummies], axis=1)
X.drop(['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope'], axis=1,inplace=True)
X.head()
```

```python
X.info()
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

model_params = {
    'svm': {
        'model': SVC(gamma='auto'),
        'params': {
            'svm__C': [1, 10, 20],
            'svm__kernel': ['rbf', 'linear']
        }
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params': {
            'random_forest__n_estimators': [1, 5, 10]
        }
    },
    'logistic_regression': {
        'model': LogisticRegression(solver='liblinear', multi_class='auto'),
        'params': {
            'logistic_regression__C': [1, 5, 10]
        }
    }
}
```

```python
scores = []

for model_name, mp in model_params.items():
    pipe = Pipeline([
        ('scaler', StandardScaler()),
        (model_name, mp['model'])
    ])
    clf = GridSearchCV(estimator = pipe, param_grid = mp['params'], cv=5, return_train_score=False)
    clf.fit(X, y)
    scores.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })
```

```python
df = pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])
df
```

> Without PCA the best model performing is logistic regression with accuracy of 82.8%, 

> parameters = solver:liblinear, multi_class='auto', C:1

```python
pca_scores = []

for model_name, mp in model_params.items():
    pipe = Pipeline([
        ('scaler', StandardScaler()),
        ('pca', PCA(.80)),
        (model_name, mp['model'])
    ])
    clf = GridSearchCV(estimator = pipe, param_grid = mp['params'], cv=5, return_train_score=False)
    clf.fit(X, y)
    pca_scores.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })
```

```python
df_pca = pd.DataFrame(pca_scores, columns=['model', 'best_score', 'best_params'])
df_pca
```

> With 80% of the information kept, we've increased accuracy of logistic regression to 83% accuracy!
